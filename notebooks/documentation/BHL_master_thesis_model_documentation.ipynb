## 2. Master thesis MOMILP notation

### 2.1. Master thesis notation pyomo_integration MILP and MOMILP notebooks

#### Weighted sum formulation
$$
    \begin{align}
        & \underset{s}{\text{min}} 
        && \bigg(\alpha \cdot \sum_{k\in \mathbb{AC_{bio}}}QB_{1,k} \cdot s_{k,1} + \beta \cdot \sum_{k\in \mathbb{AC_{c}}}C_{1,k}\cdot s_{k,1} \bigg) \\
        & \text{s.t.}   && \sum_{k\in \mathbb{AC}} A_{eq,p,k}s_{k,1} = f_{eq,p} &&& \forall p \in \mathbb{EX_{eq}} \\

        &               && f_{ineq,lb,r} \leq \sum_{k \in \mathbb{AC}} A_{ineq,r,k}s_{k,1} \leq f_{ineq,ub,r} &&& \forall r \in \mathbb{EX_{ineq}} \\
        &               && lb_k \cdot y_k \leq s_k \leq ub_k \cdot y_k &&& \forall k \in \mathbb{AC_{b}} \\
        &               && 0 \leq s_l \leq \infty, \; &&& l \in \mathbb{AC} \setminus \{\mathbb{AC_{con}} \cap \mathbb{AC_{b}}\}  \\
    \end{align}
$$
#### Sets from MA 
Missing from notation 'A_eq_index' and 'A_ineq_index'.
$$
    \begin{aligned}
        &\mathbb{AC} \in \mathbb{Z} : &&\text{activities / processes ids (variable dimension) in the technosphere matrix} \\
        &\mathbb{AC_{bio}} \subset \mathbb{AC} : &&\text{activities / processes ids in the biosphere matrix(variable dimension)} \\
        &\mathbb{AC_{con}} \subset \big\{ \mathbb{AC} \setminus \{\mathbb{AC_b}\} \big\} : &&\text{activities / processes ids which are constrained other then $s\geq$ 0 (variable dimension)} \\
        &\mathbb{AC_{b}} \subset \mathbb{AC} : &&\text{activities / processes ids can be turned on and off, meaning integer constrained $y\cdot lb \leq s\leq ub \cdot y$ 0 (variable dimension)} \\
        &\mathbb{EX_{eq}} \subset \mathbb{EX}:  &&\text{exchange ids that are equality constraints for the technosphere demand }  A \cdot s : f \\
        &\mathbb{EX_{ineq}} \subset \mathbb{EX}:  &&\text{exchange ids that are inequality constraints for the technosphere demand } f_{lb} \leq A \cdot s \leq f_{ub} \\
        &\mathbb{AC_{c}} \subset \mathbb{AC} : &&\text{activities / processes ids in the cost matrix(variable dimension)} \\
    \end{aligned}
$$

#### Parameters
$$
    \begin{aligned}
        &A_{eq} = A_{eq,m,k} &&\forall m \in \mathbb{EX_{eq}}, k \in \mathbb{AC} &&&= \text{technosphere flow factor of the equality constraints}  \\
        &A_{ineq} = A_{ineq,n,k} &&\forall n \in \mathbb{EX_{ineq}}, k \in \mathbb{AC} &&&= \text{technosphere flow factor of the inequality constraints}  \\
        &f_{eq} = f_{eq,m,1} &&\forall m \in \mathbb{EX_{eq}} &&&= \text{demand for the equality constraints}  \\
        &f_{ineq,lb} = f_{ineq,lb,m,1} &&\forall n \in \mathbb{EX_{ineq}} &&&= \text{lower bound of the demand for the inequality constraints}  \\
        &f_{ineq,ub} = f_{ineq,ub,n,1} &&\forall n \in \mathbb{EX_{ineq}} &&&= \text{upper bound of the demand for the inequality constraint }  \\        
        &lb = lb_{k,1} &&\forall k \in \mathbb{AC_{con}} &&&= \text{lower bound of the scaling variables}  \\
        &ub = ub_{k,1} &&\forall k \in \mathbb{AC_{con}} &&&= \text{upper bound of the scaling variables}  \\
        &lb_b = lb_{b,k,1} &&\forall k \in \mathbb{AC_{b}} &&&= \text{lower bound of the scaling variables which have binary activated bounds}  \\
        &ub_b = ub_{b,k,1} &&\forall k \in \mathbb{AC_{b}} &&&= \text{upper bound of the scaling variables which have binary activated bounds}  \\
        &C = C_{1,k} &&\forall k \in \mathbb{AC}_c &&&= \text{costs for activitities}  \\
        &QB = QB_{1,k} &&\forall k \in \mathbb{AC}_{bio} &&&= \text{environmental impact for activitities}  \\
        &\alpha > 0 && \in [0;1] &&&= \text{weighted sum coefficient for objective I}  \\
        &\beta > 0 && \in [0;1] &&&= \text{weighted sum coefficient for objective II}  \\
    \end{aligned}
$$

#### Variables
$s \in \mathbb{R}$: Continuous scaling variable, setting how much of each process is needed to satisfy the demand

**$$ s_k  \ \ \ \ \forall k \in  \mathbb{AC}$$**


$$
    \begin{aligned}
        &0 \leq s_{k} \leq \inf  &&\forall k \in \{ \mathbb{AC} \setminus \{\mathbb{AC_{con}} \cap \mathbb{AC_{b}}\} \} \\
        &lb_{k} \leq s_{k} \leq ub_{k} \ \ \ \ &&\forall k \in \mathbb{AC_{con}} \\
    \end{aligned}
$$

$y \in \{0,1\}$: binary bound activation varibale, used for all plant processes which can either be built/shut down completely or the opposite

**$$ y_k  \ \ \ \ \forall k \in  \mathbb{AC_b}$$**
#### Objective
##### Bi-Objective formulation

The environmental matrix multiplication $Q\cdot B$ is calculated outside the model:

$
    QB = \sum_{j\in \mathbb{EF}}Q_{1,j}B_{j,k}
$

Minimize the environmental impact of two categories

$
    \underset{s}{\text{min}} = \bigg(\sum_{k\in \mathbb{AC_{bio}}}QB_{1,k}\cdot s_{k,1}, \ \ \
     \sum_{k\in \mathbb{AC_{c}}}C_{1,k}\cdot s_{k,1} \bigg)
$

##### Weighted sum formulation
Minimize a single objective, by combining the two objectives with weight coefficients

$
    \underset{s}{\text{min}} = \bigg(\alpha \cdot \sum_{k\in \mathbb{AC_{bio}}}QB_{1,k} \cdot s_{k,1} +
     \beta \cdot \sum_{k\in \mathbb{AC_{c}}}C_{1,k}\cdot s_{k,1} \bigg)
$
#### Constraints
##### Equality constraints
$
    \begin{aligned}
        \sum_{k\in \mathbb{AC}} A_{eq,p,k}s_{k,1} = f_{eq,p} \ \ \ \ \forall p \in \mathbb{EX_{eq}}
    \end{aligned}
$
##### Inequality constraints
$
    \begin{aligned}
        f_{ineq,lb,r} \leq \sum_{k \in \mathbb{AC}} A_{ineq,r,k}s_{k,1} \leq f_{ineq,ub,r} \ \ \ \ \forall r \in \mathbb{EX_{ineq}}
    \end{aligned}
$
##### Binary inequality bound for $s$
$
        lb_k \cdot y_k \leq s_k \leq ub_k \cdot y_k \ \ \ \ \forall k \in \mathbb{AC_{b}}
$

equivalent to:

$
        s_k - lb_{b,k} \cdot y_k \geq 0 \\
        s_k - ub_{b,k} \cdot y_k \leq 0 \\
$

### 2.2. Master thesis pyomo MOMILP formulation
Based on the code in `pyomo_models.py` from the `MOMILPmodelPyomo` class.

#### Objectives
$$
\begin{align*}
    \text{obj\_1}: \quad & \text{min} \quad \sum_{k \in \mathbb{AC_{bio}}} QB_{k} \cdot s_{k} \\
    \text{obj\_2}: \quad & \text{min} \quad \sum_{k \in \mathbb{AC_{c}}} C_{k} \cdot s_{k} \\
    \text{objective}: \quad & \text{min} \quad \ \alpha \cdot \sum_{k \in \mathbb{AC_{bio}}} QB_{k} \cdot s_{k} + \beta \cdot \sum_{k \in \mathbb{AC_{c}}} C_{k} \cdot s_{k}.
\end{align*}
$$

#### **Constraints**
$$
\begin{align*}
    &\text{eq\_constraints}: \quad && \sum_{k \in \mathbb{AC}} A_{eq,p,k} s_{k,1} = f_{eq,p} \quad \forall p \in \mathbb{EX_{eq}}. \\
    &&&\text{There is an if statement in the summation only AC if (p,k) in A\_eq\_index} \\
    &\text{ineq\_constraints}: \quad && f_{ineq,lb,r} \leq \sum_{k \in \mathbb{AC}} A_{ineq,r,k} s_{k,1} \leq f_{ineq,ub,r} \quad \forall r \in \mathbb{EX_{ineq}}. \\
    &&&\text{There is an if statement in the summation only AC if (r,k) in A\_ineq\_index} \\
    &\text{binary\_ineq\_constraint\_lb}: \quad && s_k - lb_{b,k} \cdot y_k \geq 0 \quad \forall k \in \mathbb{AC_{b}}. \\
    &\text{binary\_ineq\_constraint\_ub}: \quad && s_k - ub_{b,k} \cdot y_k \leq 0 \quad \forall k \in \mathbb{AC_{b}}.
\end{align*}
$$

#### **Variables**
$$
\begin{align*}
    \text{s[k]} & : s_k && \text{Continuous scaling variable for each process.} \quad k \in \mathbb{AC}\\
    \text{y[k]} & : y_k && \text{Binary activation variable for processes.} \quad y_k \in \{0,1\} \quad k \in \mathbb{AC_b}.
\end{align*}
$$

##### Variable bounds
The variable bounds are specified in the pyomo variable definition
$$
    \begin{aligned}
        &0 \leq s_{k} \leq \inf  &&\forall k \in \{ \mathbb{AC} \setminus \{\mathbb{AC_{con}} \cap \mathbb{AC_{b}}\} \} \\
        &lb_{k} \leq s_{k} \leq ub_{k} \ \ \ \ &&\forall k \in \mathbb{AC_{con}} \\
    \end{aligned}
$$

#### **Parameters**
$$
\begin{align*}
    \text{A\_eq[p, k]} & : A_{eq,p,k} && \text{Technosphere flow factor for equality constraints.} \\
    \text{A\_ineq[r, k]} & : A_{ineq,r,k} && \text{Technosphere flow factor for inequality constraints.} \\
    \text{f\_eq[p]} & : f_{eq,p} && \text{Demand for equality constraints.} \\
    \text{f\_ineq\_lb[r], f\_ineq\_ub[r]} & : f_{ineq,lb,r}, f_{ineq,ub,r} && \text{Bounds for inequality constraints.} \\
    \text{lb[k], ub[k]} & : lb_k, ub_k && \text{Lower and upper bounds for scaling variables.} \\
    \text{lb\_b[k], ub\_b[k]} & : lb_{b,k}, ub_{b,k} && \text{Bounds for binary activated variables.} \\
    \text{QB[k]} & : QB_{1,k} && \text{Environmental impact coefficients.} \\
    \text{C[k]} & : C_{1,k} && \text{Cost coefficients.} \\
    \text{alpha, beta} & : \alpha, \beta && \text{Weighted sum coefficients.}
\end{align*}
$$

#### **Sets**
$$
\begin{align*}
    \text{AC} & : \mathbb{AC} && \text{Activities/processes IDs.} \\
    \text{AC\_bio} & : \mathbb{AC_{bio}} \in \mathbb{AC} && \text{Activities/processes in the biosphere.} \\
    \text{AC\_con} & : \mathbb{AC_{con}} \in \mathbb{AC} && \text{Constrained activities other than } s \geq 0. \\
    \text{AC\_b} & : \mathbb{AC_{b}} \in \mathbb{AC} && \text{Activities with binary activation.} \\
    \text{AC\_c} & : \mathbb{AC_{c}} && \text{Activities in the cost matrix.} \\
    \text{EX\_eq} & : \mathbb{EX_{eq}} && \text{Exchange IDs for equality constraints.} \\
    \text{EX\_ineq} & : \mathbb{EX_{ineq}} && \text{Exchange IDs for inequality constraints.} \\
    \text{A\_eq\_index} & : \text{not defined} && \text{Index of technosphere coefficient for the equality constraints, format: (exchange,process)} \\
    \text{A\_ineq\_index} & : \text{not defined} && \text{Index of technosphere coefficient for the inequality constraints, format: (exchange,process)}
\end{align*}
$$
